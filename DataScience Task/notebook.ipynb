<h3>1) Import dataset</h3>

```python
import pandas as pd

dataset=pd.read_csv('drinkMenu.csv')
dataset.columns = dataset.columns.str.replace(' ', '') #format headers

dataset.head(3)
```

```python
dataset.shape
```

<h3>2) Data preparation</h3>

<h5> There are no duplicates to remove we don't want to mess up our data</h5>

```python
#check for nulls
dataset.isnull().any()
```

<h5>Fill the Caffeine cells with the average val according to the Beverage_category</h5>

```python
dataset["Caffeine(mg)"] = dataset.groupby("Beverage_category").transform(lambda x: x.fillna(x.mean()))
```

```python
dataset.isnull().any()
```

<h5> We can get rid of TransFat(g) ,SaturatedFat(g) since we have the column TotalFat(g)</h5>

```python
dataset=dataset.drop(['TransFat(g)', 'SaturatedFat(g)'], axis=1)
dataset.shape
```

<h3>3) Data Visualisations </h3>

```python
from matplotlib import pyplot as plt
```

<h5> ‚óè Which drink has the highest calories from the dataset?</h5>

```python

drinks = dataset['Beverage']
calories = dataset['Calories']
 
# Figure Size
fig = plt.figure(figsize =(15, 9))
plt.xticks(rotation='vertical')
# Horizontal Bar Plot
plt.bar(drinks, calories)
 
# Add title and axis names
plt.title('Drinks - Calories')
plt.xlabel('Drinks')
plt.ylabel('Calories')
 
# Show Plot
plt.show()

```

<h5>Answer: Drink with the highest calorie is White Chocolate Mocha (Without Whipped Cream) </h5>

<h5> ‚óè What is the highest sugar drink ?</h5>

```python
sugars = dataset['Sugars(g)']

# Figure Size
fig = plt.figure(figsize =(17, 12))

plt.xticks(rotation='vertical')
# Horizontal Bar Plot
plt.bar(drinks, sugars)

#valuelabel(drinks, sugars)

# Add title and axis names
plt.title('Drinks - Sugars')
plt.xlabel('Drinks')
plt.ylabel('Sugars')
plt.grid()

# Show Plot
plt.show()
```

<h5>Answer: Drink with the highest sugar is Java chip (Without Whipped Cream)</h5>
